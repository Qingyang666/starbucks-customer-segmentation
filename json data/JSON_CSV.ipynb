#### Output Transcript Data To CSV

```python
# read in the transcript data 
transcript = pd.read_json('transcript.json', orient = 'records', lines = True)
```

```python
transcript.head()
```

```python
transcript.tail()
```

```python
# Add column `value_type`
transcript['value_type'] = transcript.value.apply(lambda x : list(x.keys())[0])

# Add column `actual_value`
transcript['actual_value'] = transcript.value.apply(lambda x : list(x.values())[0])
```

```python
transcript.head()
```

```python
transcript.tail()
```

```python
# write out as csv file
transcript.to_csv('transcript.csv')
```
